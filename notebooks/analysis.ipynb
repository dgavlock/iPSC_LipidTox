```python
import pandas as pd 
import numpy as np
import re
```

```python
df = pd.read_csv('../data/processed/processed1.csv')
df
```

```python
import matplotlib.pyplot as plt

# Select relevant columns
lipid_area_col = "Lipid - Lipid Area [µm²] - Mean per Well"
condition_col = "Condition"

# Group by condition and calculate the mean lipid area
lipid_area_by_condition = df.groupby(condition_col)[lipid_area_col].mean()

# Plot the results
plt.figure(figsize=(10, 6))
lipid_area_by_condition.plot(kind='bar')
plt.xlabel("Condition")
plt.ylabel("Mean Lipid Area (µm²) per Well")
plt.title("Mean Lipid Area by Condition")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
